**Mengimpor library Pandas dan Numpy & Membaca File Data dari Dataset "Global Missing Migrants Dataset"**

```python
import numpy as np
import pandas as pd

#membaca file dari dataset
dataFrame = pd.read_csv('Global Missing Migrants Dataset.csv')

#menampilkan data dalam dataset
dataFrame
```

**Menampilkan Rata-Rata,  standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka**

```python
cNumerik = dataFrame.select_dtypes(include = np.number)
#untuk menghitung nilai rata-rata
rata_rata = cNumerik.mean()
#untuk menghitung nilai standar deviasi
standar_deviasi = cNumerik.std()
#untuk menghitung nilai min dan nilai maksimal
nilai_min = cNumerik.min()
nilai_maks = cNumerik.max()
#untuk menghitung nilai quartil1, nilai quartil2, dan quartil3
Q1 = cNumerik.quantile(25/100)
Q2 = cNumerik.quantile(50/100)
Q3 = cNumerik.quantile(75/100)
#untuk menampilkan hasil dari perhitungan diatas
informasi_hasil = pd.DataFrame({
    "Rata-Rata" : rata_rata,
    "Standar Deviasi" : standar_deviasi,
    "Nilai Maks" : nilai_maks,
    "Nilai Min" : nilai_min,
    "Q1" : Q1,
    "Q2" : Q2,
    "Q3" : Q3,
})
print(informasi_hasil.to_string())
```

**Menampilkan Jumlah Record, Jumlah Atribut, Nama Atribut, Jumlah Atribut Bertipe Angka(Numerik), Serta Menampilkan Informasi Tersebut dalam Dictionary dengan Key Sesuai Namanya**

```python
#untuk menghitung jumlah record
len(dataFrame)
#untuk menghitung jumlah atribut
len(dataFrame.columns)
#untuk menampilkan nama atribut dalam bentuk list
dataFrame.columns.tolist()
#untuk menampilkan informasi dalam bentuk dictionary
len(dataFrame.select_dtypes(include= np.number).columns)
informasi_dict = {
    "Jumlah Record" : len(dataFrame),
    "Jumlah Atribut" : len(dataFrame.columns),
    "Jumlah Atribut dengan Tipe Angka " : len(dataFrame.select_dtypes(include=['number']).columns),
    "Nama Atribut" : dataFrame.columns.tolist()
}
#untuk menampilkan informasi dalam design table
print("|********************************************************************|")
print("|*****     DATA      ************************       HASIL      ******|")
print("|********************************************************************|")
#menampilkan informasi berdasarkkan data yang dimasukkan didalam ctionary
key_max = max(len(key) for key in informasi_dict.keys())
for key, value in informasi_dict.items():
    if key == "Nama Atribut":
        for atribut in value[1:]:
            print("|{:<31}|-{:<35}|".format("", atribut))
    else:
        print("|{:<31}|-{:<35}|".format(key, value))
        
        print("|********************************************************************|")
```

**Menampilkan Jumlah Nilai Dari Semua atribut yang bernilai object**

```python
cObjek = dataFrame.select_dtypes(include="object")
jumlah_col={}
#untuk menghitung jumlah nilai unik setiap atribut dan ditampilkan dalam dictionary
for atribut in cObjek.columns:
    
   #print(f"Jumlah nilai dalam setiap atribut '{atribut}':")
   #print(dataFrame[atribut].value_counts())
   #print("\n")
   jumlah_nilai = len(cObjek[atribut].value_counts())
   jumlah_col[atribut] = jumlah_nilai
   print("|********************************************************************|")
   print("|*****           ATRIBUT        **********         JUMLAH       *****|")
   for atribut, jumlah in jumlah_col.items():
       print("|{:<34}|                 {:<17}|".format(atribut, jumlah_nilai))
```
