**MENAMPILKAN GRAFIK MENGGUNAKAN PIE CHART**

```python
# Hitung jumlah kejadian per wilayah geografis dalam variabel
region_counts = dataFrame['Region of Incident'].value_counts()
# mengatur Persentase kejadian
explode = (0.3, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3, 0.3)
# menentukan warna yang berbeda untuk setiap wilayah
colors = plt.cm.Paired(range(len(region_counts)))
# untuk membuat grafik piechart
plt.pie(region_counts, labels = region_counts.index, colors = colors, autopct = '%1.1f%%', startangle=180)
# Judul grafik
plt.title('Grafik Persentase Kejadian Berdasarkan Wilayah Geografisnya ', pad=50)
# mengatur aspek grafik agar menjadi lingkaran
plt.axis('equal')
# menampilkan grafik
plt.tight_layout()
plt.legend(region_counts.index, title = "Region", loc="best", bbox_to_anchor=(1.5, 0.8))
plt.show()
```

**MENAMPILKAN DATA MENGGUNAKAN LINE PLOT**

```python
#variabel "year_counts" untuk mengisi daftar kolom dari dataset yang akan ditampilkan
year_counts = dataFrame.groupby('Incident year')['Total Number of Dead and Missing'].mean()

#untuk membuat figur gambar 
plt.figure(figsize=(12, 6))

#untuk membuat plot garis yang akan ditampilkan dari sumbu x dan sumbu y
plt.plot(year_counts.index, year_counts.values,  marker='o', linestyle='-', color='green', label='Tahun Kejadian')
plt.title('Rata-Rata Jumlah Kejadian Per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kejadian')
plt.xticks(rotation  =50)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### KETERANGAN :
- Data diatas menggambarkan rata-rata jumlah kejadian (Total Number of Dead and Missing) per tahun (Incident year).
- Grafik menunjukkan tren umum bahwa rata-rata jumlah kejadian seiring berjalannya waktu. Hal ini dapat mengindikasikan bahwa masalah yang terkait dengan kehilangan dan kematian migran mungkin semakin mendalam atau mendapatkan perhatian lebih besar dalam beberapa tahun terakhir.
- Terdapat lonjakan tajam dalam rata-rata jumlah kejadian pada tahun tertentu
- Beberapa tahun mengalami peningkatan dalam jumlah kejadian, sementara tahun lain mengalami penurunan atau tetap relatif stabil.
